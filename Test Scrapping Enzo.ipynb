```python
# Liste des 'NOCs'
url = "https://en.wikipedia.org/wiki/List_of_IOC_country_codes#Current_NOCs"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = urllib.request.Request(url,headers=hdr)
source = urllib.request.urlopen(req).read()
soup = bs.BeautifulSoup(source,'lxml')

table = soup.find('table')

list_of_NOCs = []

table_rows = table.find_all('tr')
for tr in table_rows:
    cols = tr.find_all('td')
    if len(cols)>0:
        list_of_NOCs.append([cols[0].text.strip(),cols[1].text.strip()])
        
df_NOCs_list = pd.DataFrame(list_of_NOCs,columns=['Code','National Olympic Committee'])

df_NOCs_list.to_excel('NOC_COUNTRY_CODES.xlsx', index=False)

df_NOCs_list.head(5)
```

```python
def country_to_continent(country_alpha3, df_NOC):
    if country_alpha3=="ENG" or country_alpha3=="SCO" or country_alpha3=="WAL":
        return 'Europe'
    elif country_alpha3=="IRI" or country_alpha3=="KOR" or country_alpha3=="TPE" or country_alpha3=="VIE" or country_alpha3=="HKG":
        return 'Asia'
    elif country_alpha3=="VEN":
        return "South America"
    else:
        country_name = df_NOC[df_NOC['Code']==country_alpha3]['National Olympic Committee'].values[0]
        country = pycountry.countries.get(name=country_name)

        country_continent_code = pc.country_alpha2_to_continent_code(country.alpha_2)
        country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    
    return country_continent_name
```

```python
url = "https://corporate.bwfbadminton.com/events/calendar/2022/all/0/-1/"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = urllib.request.Request(url,headers=hdr)
source = urllib.request.urlopen(req).read()
soup = bs.BeautifulSoup(source,'lxml')

tables = soup.find_all('table')

tournaments_list = []

df_NOC = pd.read_excel('NOC_COUNTRY_CODES.xlsx')

month = 0
for table in tables:
    month+=1
    table_rows = table.find_all('tr')
    for tr in table_rows:
        cols = tr.find_all('td')
        if len(cols)>6 and not "Junior" in cols[5].text:
            # 0: Week, 1: Country, # 2: Dates, # 3: Tournament, #5: Category, # 6: City
            try:
                continent = country_to_continent(cols[1].text.strip(),df_NOC)
            except: 
                continent = None
            tournaments_list.append([str(month),cols[0].text.strip(),cols[2].text.strip(),continent,cols[1].text.strip(),cols[6].text.strip(),cols[3].text.strip(),cols[5].text.strip()])
        
df_tournaments_list = pd.DataFrame(tournaments_list,columns=['Month','Week','Dates','Continent','Country','City','Tournament','Category'])

df_tournaments_list.to_excel('2022_BWF_TOURNAMENTS_LIST.xlsx',index=False)

df_tournaments_list.head(5)
```

```python
#Liste des tournois avec les dates clÃ©s
url = "https://corporate.bwfbadminton.com/events/calendar/2022/all/0/-1/"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = urllib.request.Request(url,headers=hdr)
source = urllib.request.urlopen(req).read()
soup = bs.BeautifulSoup(source,'lxml')

tables = soup.find_all('ul', attrs={'class':'list-unstyled bwf-ul'})

key_dates = [] #Entry Deadline, World Ranking Date, M & Q Report, Seeding Date, Withdrawal Deadline, Draw Date, Tournament Link, Website, Updated Date

for table in tables:
    key_dates_tournament = ["","","","","","","","",""]
    table_rows = table.find_all('li')
    for j in range(len(table_rows)):
        if "Entry Deadline" in table_rows[j].text:
            date = table_rows[j].text.split(':')[1].strip()
            key_dates_tournament[0]= date
        elif "World Ranking Date" in table_rows[j].text:
            date = table_rows[j].text.split(':')[1].strip()
            key_dates_tournament[1]= date
        elif "M & Q Report" in table_rows[j].text:
            date = table_rows[j].text.split(':')[1].strip()
            key_dates_tournament[2]= date
        elif "Seeding Date" in table_rows[j].text:
            date = table_rows[j].text.split(':')[1].strip()
            key_dates_tournament[3]= date
        elif "Withdrawal Deadline" in table_rows[j].text:
            date = table_rows[j].text.split(':')[1].strip()
            key_dates_tournament[4]= date
        elif "Draw Date" in table_rows[j].text:
            date = table_rows[j].text.split(':')[1].strip()
            key_dates_tournament[5]= date
        elif "Tournament Link" in table_rows[j].text:
            key_dates_tournament[6]= table_rows[j].find('a').get('href')
        elif "Website" in table_rows[j].text:
            key_dates_tournament[7]= table_rows[j].find('a').get('href')
        elif "Updated date" in table_rows[j].text:
            date = table_rows[j].text.split(':')[1].strip()
            key_dates_tournament[8]= date
    key_dates.append(key_dates_tournament)

df_tournaments_key_dates = pd.DataFrame(key_dates,columns=['Entry Deadline', 'World Ranking Date', 'M & Q Report', 'Seeding Date', 'Withdrawal Deadline', 'Draw Date', 'Tournament Link', 'Website', 'Updated Date'])

df_tournaments = pd.read_excel('2022_BWF_ALL_TOURNAMENTS_LIST.xlsx')

df_tournaments_key_dates = pd.concat([df_tournaments,df_tournaments_key_dates],axis=1)

df_tournaments_key_dates.to_excel('2022_BWF_ALL_TOURNAMENTS_LIST_WITH_KEY_DATES.xlsx',index=False)

df_tournaments_key_dates
```

```python
df_distributions = pd.read_excel('BWF_TOURNAMENTS_POINTS_DISTRIBUTION.xlsx')

for i in range(len(df_tournaments)):
    
    #On dÃ©tecte les tournois par Ã©quipe
    if "Team" in df_tournaments['Category'].iloc[i]:
        df_tournaments.loc[i,'Team']="1"
    else:
        df_tournaments.loc[i,'Team']="0"
        
    #On fait le lien entre les deux tableaux
    if df_tournaments['Category'].iloc[i]=='HSBC BWF World Tour Finals':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='BWF World Tour Finals'].iloc[0,3:].values

    elif df_tournaments['Category'].iloc[i]=='HSBC BWF World Tour Super 1000':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='BWF World Tour Super 1000'].iloc[0,3:].values

    elif df_tournaments['Category'].iloc[i]=='HSBC BWF World Tour Super 750':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='BWF World Tour Super 750'].iloc[0,3:].values

    elif df_tournaments['Category'].iloc[i]=='HSBC BWF World Tour Super 500':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='BWF World Tour Super 500'].iloc[0,3:].values

    elif df_tournaments['Category'].iloc[i]=='HSBC BWF World Tour Super 300':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='BWF World Tour Super 300'].iloc[0,3:].values
    
    elif df_tournaments['Category'].iloc[i]=='BWF Tour Super 100':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='BWF World Tour Super 100'].iloc[0,3:].values
    
    elif df_tournaments['Category'].iloc[i]=='International Challenge':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='International Challenge'].iloc[0,3:].values

    elif df_tournaments['Category'].iloc[i]=='International Series':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='International Series'].iloc[0,3:].values

    elif df_tournaments['Category'].iloc[i]=='Future Series':
        df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='Future Series'].iloc[0,3:].values

    elif df_tournaments['Category'].iloc[i]=='Grade 1 - Individual Tournaments':
        if "World Championships" in df_tournaments['Tournament'].iloc[i]:
            df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='BWF World Championships'].iloc[0,3:].values
        elif "Olympic Games" in df_tournaments['Tournament'].iloc[i]:
            df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='Olympic Games'].iloc[0,3:].values
            
    elif df_tournaments['Category'].iloc[i]=='Continental Individual Games' or df_tournaments['Category'].iloc[i]=='Multi-Sport Games' or df_tournaments['Category'].iloc[i]=='Continental Individual Championships':
        continent = df_tournaments['Continent'].iloc[i]
        if continent=='Europe':
            df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='European Championships'].iloc[0,3:].values

        elif continent=='Asia':
            df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='Asian Championships'].iloc[0,3:].values

        elif continent=='Africa':
            df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='African Championships'].iloc[0,3:].values

        elif continent=='Oceania':
            df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='Oceania Championships'].iloc[0,3:].values

        elif continent=='South America' or continent=='North America':
            df_tournaments.iloc[i,9:] = df_distributions[df_distributions['Name']=='Pan American Championships'].iloc[0,3:].values

df_tournaments.to_excel('2022_BWF_TOURNAMENTS_LIST_WITH_DISTRIBUTION.xlsx',index=False)
        
df_tournaments.head(5)
```

```python
#Simple Homme
df_french_players = pd.read_excel('./FRANCE_PLAYERS/MEN_SINGLES_FRANCE_PLAYERS.xlsx')

df_french_players = df_french_players[df_french_players['Points']>=1000]

df_french_players
```

```python
#Fonction qui va catÃ©goriser les tournois
def get_category_tournaments(df_player_tournaments_results,df_tournaments_list):
    #On regarde si les tournois sont par equipe
    for i in range(len(df_player_tournaments_results)):
        df_tournament_details = df_tournaments_list[df_tournaments_list['Tournament']==df_player_tournaments_results['Tournament'].iloc[i]]
        df_player_tournaments_results.loc[i,'Team'] = df_tournament_details['Team'].values[0]
    
        if df_tournament_details['Category'].iloc[0]=='Continental Individual Games' or df_tournament_details['Category'].iloc[0]=='Multi-Sport Games':
            df_player_tournaments_results.loc[i,'Continental Games'] = 1
        else:
            df_player_tournaments_results.loc[i,'Continental Games'] = 0

        if df_tournament_details['Category'].iloc[0]=='Continental Individual Championships':
            df_player_tournaments_results.loc[i,'Continental Championships'] = 1
        else:
            df_player_tournaments_results.loc[i,'Continental Championships'] = 0

        if "World Championships" in df_tournament_details['Tournament'].iloc[0]:
            df_player_tournaments_results.loc[i,'World Championships'] = 1
        else:
            df_player_tournaments_results.loc[i,'World Championships'] = 0

    df_player_tournaments_results['Team'] = df_player_tournaments_results['Team'].astype('int')
    df_player_tournaments_results['Continental Games'] = df_player_tournaments_results['Continental Games'].astype('int')
    df_player_tournaments_results['Continental Championships'] = df_player_tournaments_results['Continental Championships'].astype('int')
    df_player_tournaments_results['World Championships'] = df_player_tournaments_results['World Championships'].astype('int')

    return df_player_tournaments_results

#Fonction qui renvoie la liste des tournois avec un attribut "Calculation"=True pour les tournois qui comptent pour le classement
def get_world_ranking(df_player_tournaments_results):
    #On les classe par ordre de points
    df_player_tournaments_results = df_player_tournaments_results.sort_values(by=['Points'], ascending=False)
    #On reset les indices
    df_player_tournaments_results.reset_index(drop=True,inplace=True)
    #Initialisation 
    df_player_tournaments_results['Calculation']=False
    df_tournaments_for_calculation = df_player_tournaments_results.iloc[0:1]
    #On parcourt tous les tournois et on vÃ©rifie qu'ils respectent les conditions pour le classement mondial
    for i in range(1,len(df_player_tournaments_results)):
        if len(df_tournaments_for_calculation)<10:
            df_tournaments_for_calculation = pd.concat([df_tournaments_for_calculation,df_player_tournaments_results.iloc[i:i+1]])
            #On regarde si il y a deux tournois par equipe
            if df_tournaments_for_calculation['Team'].sum()>1:
                index_team = df_tournaments_for_calculation.index[df_tournaments_for_calculation['Team']==1].tolist()
                if df_tournaments_for_calculation.loc[index_team[0],"Points"]>=df_tournaments_for_calculation.loc[index_team[1],"Points"]:
                    print(df_tournaments_for_calculation.loc[index_team[0],"Points"])
                    df_tournaments_for_calculation.drop(index_team[1], inplace=True)
                else:
                    df_tournaments_for_calculation.drop(index_team[0], inplace=True)

            #On regarde si il y a deux jeux continentaux
            if df_tournaments_for_calculation['Continental Games'].sum()>1:
                index_cont_games = df_tournaments_for_calculation.index[df_tournaments_for_calculation['Continental Games']==1].tolist()
                if df_tournaments_for_calculation.loc[index_cont_games[0],"Points"]>=df_tournaments_for_calculation.loc[index_cont_games[1],"Points"]:
                    df_tournaments_for_calculation.drop(index_cont_games[1], inplace=True)
                else:
                    df_tournaments_for_calculation.drop(index_cont_games[0], inplace=True)

            #On regarde si il y a deux championnats continentaux
            if df_tournaments_for_calculation['Continental Championships'].sum()>1:
                index_cont_champ = df_tournaments_for_calculation.index[df_tournaments_for_calculation['Continental Championships']==1].tolist()
                if df_tournaments_for_calculation.loc[index_cont_champ[0],"Points"]>=df_tournaments_for_calculation.loc[index_cont_champ[1],"Points"]:
                    df_tournaments_for_calculation.drop(index_cont_champ[1], inplace=True)
                else:
                    df_tournaments_for_calculation.drop(index_cont_champ[0], inplace=True)

            #On regarde si il y a deux championnats du monde       
            if df_tournaments_for_calculation['World Championships'].sum()>1:
                index_world_champ = df_tournaments_for_calculation.index[df_tournaments_for_calculation['World Championships']==1].tolist()
                if df_tournaments_for_calculation.loc[index_world_champ[0],"Points"]>=df_tournaments_for_calculation.loc[index_world_champ[1],"Points"]:
                    df_tournaments_for_calculation.drop(index_world_champ[1], inplace=True)
                else:
                    df_tournaments_for_calculation.drop(index_world_champ[0], inplace=True)

    #On rÃ©cupÃ¨re les indices des tournois pour le calcul du classement
    index_tournaments_for_calculation = df_tournaments_for_calculation.index
    for idx in index_tournaments_for_calculation:
        df_player_tournaments_results.loc[idx,"Calculation"] = True
        
    return df_player_tournaments_results
        

    
```
